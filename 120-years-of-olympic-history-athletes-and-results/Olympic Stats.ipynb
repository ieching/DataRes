```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("athlete_events.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
def convert_medal(medal):
    if medal == "Gold":
        return 1
    elif medal == "Silver":
        return 2
    elif medal == "Bronze":
        return 3
    elif np.isnan(medal):
        return 0
```

```python
df["Medal"] = df["Medal"].apply(convert_medal)
```

```python
df.head()
```

```python
df[df["Sport"] == "Basketball"]
```

```python
numeric_df = pd.DataFrame([df[df["Sport"] == "Basketball"]["Age"], df[df["Sport"] == "Basketball"]["Height"], df[df["Sport"] == "Basketball"]["Weight"], df[df["Sport"] == "Basketball"]["Year"], df[df["Sport"] == "Basketball"]["Medal"]]).transpose()
```

```python
numeric_df.head()
```

```python
sns.pairplot(numeric_df)
```

```python
teams = df["Team"].unique()
medals = []

for team in teams:
    total = df[df["Team"] == team]["Medal"].sum()
    medals.append(total)
```

```python
team_df = pd.DataFrame([teams, medals]).transpose()
team_df.columns = ['Team', 'Total Medals']
```

```python
team_df.head()
```

```python
team_df = team_df.sort_values(by="Total Medals", ascending=False)
```
